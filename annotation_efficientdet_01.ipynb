```python
def format_labels(filename, label, path): #label = TEST, TRAIN VALIDATION
    height = 512
    width = 512
    df = pd.DataFrame()

    for file in filename:
        tree = ET.parse(path + file + '.xml') #change here
        root = tree.getroot()
        for i in root.findall("object"):
            row = [label, f"/content/efficientDet/image_subset/{file}.jpg", "pool"] #change here #TRAIN VALIDATION TEST UNASSIGNED

            coords = i.find("bndbox")
            xmin = float(coords.find("xmin").text)/width
            ymin = float(coords.find("ymin").text)/height
            xmax = float(coords.find("xmax").text)/width
            ymax = float(coords.find("ymax").text)/height

            row.extend([xmin, ymin, "", "", xmax, ymax, "", ""])

            df = df.append([row])
            
    print(f'Length of df: {len(df)}')
    df_formatted = df.copy()
    
    return(df_formatted)
```

```python
df_train = format_labels(train, "TRAIN", path_annots_pascal)
df_val = format_labels(val, "VALIDATION", path_annots_pascal)
df_test = format_labels(test, "TEST", path_annots_pascal)
```

```python
csv_data = pd.DataFrame()
frames = [df_train, df_val, df_test]
df_all = csv_data.append(frames)

print(f'Length of df: {len(df_all)}')

df_all.to_csv('annotations.csv', header=False, index=False)
```
